```python
import pandas as pd
from sklearn.metrics import roc_curve, roc_auc_score
from matplotlib import pyplot as plt
```

```python
df_task2 = pd.read_csv("../analysis/2/task2_CnnOnly_validate_2025-10-28 14:52:54.008621.csv")
df_task3 = pd.read_csv("../analysis/3/task3_CnnOnly_validate_2025-10-28 14:58:41.258842.csv")
df_task4 = pd.read_csv("../analysis/4/task4_CnnOnly_validate_2025-10-28 15:05:11.569730.csv")
df_task5 = pd.read_csv("../analysis/5/task5_CnnOnly_validate_2025-10-28 15:14:37.034557.csv")
df_task6 = pd.read_csv("../analysis/6/task6_CnnOnly_validate_2025-10-28 15:25:55.115154.csv")
df_task7 = pd.read_csv("../analysis/7/task7_CnnOnly_validate_2025-10-28 15:33:34.406800.csv")
df_task8 = pd.read_csv("../analysis/8/task8_CnnOnly_validate_2025-10-28 15:42:24.928313.csv")

dataframes = {
    2: df_task2,
    3: df_task3,
    4: df_task2,
    5: df_task3,
    6: df_task2,
    7: df_task3,
    8: df_task2,
}
```

```python
for i in range(2,9):
    df = dataframes[i]
    y_true = df["target"]
    y_score = df["confidence"]

    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    auc = roc_auc_score(y_true, y_score)

    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    plt.plot([0,1], [0,1], "k--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"Curva ROC task {i}")
    plt.legend(loc="lower right")
    plt.savefig(f"ROC_curve_task{i}.png", dpi=300, bbox_inches="tight")
    plt.close()
```
